```python
from sklearn.metrics import mean_squared_error
import math
import pandas as pd
y_actual = [1,2,3,4,5]
y_predicted = [1.6,2.5,2.9,3,4.1]
 
MSE = mean_squared_error(y_actual, y_predicted)
 
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

```python
import numpy as np
import pprint as pp
rmse_pd = pd.DataFrame(columns=['data_label', 'ses', 'svr', 'nn', 'rf', 'gpr', 'arima', 'prophet', 'silverkite', 'adarel'])
file_names= ['2021DataSet1.csv', '2021DataSet2.csv', '2021DataSet3.csv', '2021DataSetSEC.csv']

def get_rmse_list(filename="2021DataSet1.csv"):
    algos = ['ses', 'svr', 'nn', 'rf', 'gpr', 'arima', 'prophet', 'silverkite', 'adarel']

    ds = pd.read_csv(filename)
    # ds = ds[~ds.isin([np.nan, np.inf, -np.inf]).any(1)]
    ds = ds.replace([np.nan, np.inf, -np.inf], 1)
    #ds = ds[np.isfinite(ds).all(1)]

    result2 = {algo: math.sqrt(mean_squared_error(ds["true value"], ds[algo])) for algo in algos}
    result2['data_label'] = filename
    
    print(result2)
    return result2
all_res = [get_rmse_list(f) for f in file_names]
pp.pprint(all_res)
rmse_pd = pd.DataFrame.from_records(all_res)


rmse_pd.head()
```
